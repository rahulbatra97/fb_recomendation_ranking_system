Data Cleaning:
Removing null values and sorting the prices column

```python
df = dataset.dropna() # deletes columns where at least 1 value is missing
df['price'] = df['price'].replace({r'£': '', r',': '', r'\)': '', r'-': '', r' ': ''}, regex=True)
#drop duplicates df.drop_duplicates()
#need to convert the price column to floats
df['price'] = df['price'].astype('float64')
df
```
